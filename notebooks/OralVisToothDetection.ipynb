```python
from ultralytics import YOLO

model = YOLO("yolov8s.pt")  # pretrained small model
model.train(
    data="/content/drive/MyDrive/OralVis-Tooth-Detection/configs/data.yaml",
    epochs=100,
    imgsz=640,
    batch=16,
    project="runs",
    name="fdiyolo",
    pretrained=True
)
```

```python
best_weights = "runs/fdiyolo/weights/best.pt"
print("Best weights saved at:", best_weights)
```

```python
best_weights = "runs/fdiyolo/weights/best.pt"
metrics = model.val(data="/content/drive/MyDrive/OralVis-Tooth-Detection/configs/data.yaml")
print(metrics.results_dict)
```

```python
model.predict(
    source="/content/drive/MyDrive/OralVis-Tooth-Detection/data/images/test",
    save=True,
    save_txt=True,
    save_conf=True,
    project="runs",
    name="predict"
)
```

```python
import shutil

# Change this to your Google Drive project folder
SAVE_DIR = "/content/drive/MyDrive/OralVis-Tooth-Detection-Results"

# Create save dir
!mkdir -p $SAVE_DIR

# Copy best model weights
shutil.copy("runs/fdiyolo5/weights/best.pt", f"{SAVE_DIR}/best.pt")

# Copy training curves & confusion matrix
shutil.copy("runs/fdiyolo5/results.png", f"{SAVE_DIR}/training_curves.png")
shutil.copy("runs/fdiyolo5/confusion_matrix.png", f"{SAVE_DIR}/confusion_matrix.png")

print("✅ Model and results saved to Google Drive:", SAVE_DIR)
```

```python
metrics = model.val(data="/content/drive/MyDrive/OralVis-Tooth-Detection/configs/data.yaml")
results = getattr(metrics, 'results_dict', None)

import json, pandas as pd

if results:
    # Save as JSON
    with open(f"{SAVE_DIR}/metrics.json", "w") as f:
        json.dump(results, f, indent=2)

    # Save as CSV
    pd.DataFrame([results]).to_csv(f"{SAVE_DIR}/metrics.csv", index=False)

print("✅ Metrics saved in JSON and CSV format")
```

```python
# Run predictions on test set
pred = model.predict(
    source="data/images/test",
    save=True,
    save_txt=True,
    save_conf=True,
    project="runs",
    name="predict"
)

# Copy predictions folder to Drive
shutil.copytree("runs/predict", f"{SAVE_DIR}/predict", dirs_exist_ok=True)
print("✅ Predictions saved to:", f"{SAVE_DIR}/predict")
```
